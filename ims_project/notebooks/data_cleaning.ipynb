Step 2: Fix dates and data types

```python
# Convert date columns to datetypes
camp_df['date'] = pd.to_datetime(camp_df['date'])
store_sales_df['date'] = pd.to_datetime(store_sales_df['date'])
cust_df['signup_date'] = pd.to_datetime(cust_df['signup_date'])
cust_df['last_purchase_date'] = pd.to_datetime(cust_df['last_purchase_date'])
web_df['date'] = pd.to_datetime(web_df['date'])

# Add useful time columns
camp_df['month'] = camp_df['date'].dt.month
camp_df['year'] = camp_df['date'].dt.year
camp_df['day_name'] = camp_df['date'].dt.day_name()
camp_df['week_num'] = camp_df['date'].dt.isocalendar().week
```
